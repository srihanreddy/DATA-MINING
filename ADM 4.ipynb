```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='mean')
```

```python
df.isnull()
```

```python
df['Age']=imp.fit_transform(df[['Age']])
```

```python
df.isnull()
```

```python
df.info()
```

```python
imp1=SimpleImputer(strategy='most_frequent')
```

```python

df['Cabin'] = imp1.fit_transform(df[['Cabin']])[:, 0]
```

```python
df.isnull()
```

```python

df_encoded = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
print(df_encoded.head())
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df_encoded['Sex'] = label_encoder.fit_transform(df_encoded['Sex'])

print(df_encoded[['Sex', 'Age', 'Fare']].head())
```
