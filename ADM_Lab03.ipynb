```python
df[['Age', 'BMI', 'Family_Income', 'Stress_Level']].hist(bins=20, figsize=(10, 8))
plt.show()

sns.boxplot(x=df['BMI'])
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].map({'Yes': 1, 'No': 0})

correlation = df.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
print(df.columns)
sns.boxplot(x=df['Smoking'])
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = ['Family_History_Diabetes', 'Smoking', 'Alcohol_Consumption']
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['BMI', 'Cholesterol_Level', 'Fasting_Blood_Sugar', 'HbA1c']])
plt.title("Boxplot of Key Variables")
plt.show()
```

```python
sns.pairplot(df[['BMI', 'Cholesterol_Level', 'Fasting_Blood_Sugar', 'HbA1c']])
plt.show()
```

```python

print(df[['Diabetes_Type', 'Fasting_Blood_Sugar']].isnull().sum())

df = df.dropna(subset=['Diabetes_Type', 'Fasting_Blood_Sugar'])

print(df['Diabetes_Type'].value_counts())

for diabetes_type in df['Diabetes_Type'].unique():
    if df[df['Diabetes_Type'] == diabetes_type]['Fasting_Blood_Sugar'].empty:
        print(f"Diabetes Type '{diabetes_type}' has no corresponding Fasting Blood Sugar values.")

sns.boxplot(x='Diabetes_Type', y='Fasting_Blood_Sugar', data=df)
plt.title("Fasting Blood Sugar Distribution by Diabetes Type")
plt.show()

sns.countplot(x='Physical_Activity_Level', hue='Prediabetes', data=df)
plt.title("Physical Activity Level and Prediabetes")
plt.show()
```

```python
sns.boxplot(x='Diabetes_Type', y='Sleep_Hours', data=df)
plt.title("Sleep Hours Distribution by Diabetes Type")
plt.show()
```

```python
sns.boxplot(x='Diabetes_Type', y='Stress_Level', data=df)
plt.title("Stress Level by Diabetes Type")
plt.show()
```
