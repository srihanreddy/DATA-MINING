```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset, display first 5 rows, and handle missing values
titanic = pd.read_csv("titanic.csv")
print("First 5 rows:")
print(titanic.head())
print("\nMissing values:")
print(titanic.isnull().sum())
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Fare'].fillna(titanic['Fare'].median(), inplace=True)
titanic.fillna(method='ffill', inplace=True)
```

```python
# 2. Calculate & visualize survival rate based on Passenger Class and Gender
survival_rate = titanic.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()
print("\nSurvival Rate by Pclass and Sex:")
print(survival_rate)
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=survival_rate)
plt.title("Survival Rate by Passenger Class and Gender")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# 3. Histogram for Age column
plt.hist(titanic['Age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

```

```python
# 4. Box plot for Fare column
sns.boxplot(y=titanic['Fare'])
plt.title("Box Plot of Fare")
plt.ylabel("Fare")
plt.show()
```
